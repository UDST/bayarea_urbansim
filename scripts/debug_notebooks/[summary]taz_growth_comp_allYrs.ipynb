```python
import pandas as pd
import os
import sys
# sys.path.insert(0, 'X:\\bayarea_urbansim\\scripts')
# from baus_analytics_tools import *
```

```python
M_drive_dir = r'M:\Data\Urban\BAUS\PBA50\explore_developer_model\runs_new_parcel_subzone_input'
```

```python
# read TAZ-SD number crosswalk
taz_f = pd.read_csv('M:/Data/GIS layers/TM1_taz/bayarea_rtaz1454_rev1_WGS84.csv',
                    usecols = ['TAZ1454', 'SUPERD'])
taz_f.drop_duplicates(inplace=True)

# read SD number-name crosswalk 
sd_name = pd.read_csv(r'X:\bayarea_urbansim\data\superdistricts.csv',
                      usecols=['number', 'name'])
sd_name.drop_duplicates(inplace=True)

# merge the two
TAZ_SD = taz_f.merge(sd_name, left_on='SUPERD', right_on='number', how='left')
TAZ_SD = TAZ_SD[['TAZ1454', 'SUPERD', 'name']]
TAZ_SD.rename(columns = {'TAZ1454': 'TAZ', 'name': 'SD_NAME'}, inplace=True)

display(TAZ_SD)
```

```python
# location of two runs: run313-P, run310-NP
runs_dir = r'X:\bayarea_urbansim\runs'
output_dir = M_drive_dir
```

```python
np_taz = TAZ_SD.copy()
for filename in os.listdir(runs_dir):
    if ('run310_taz_summaries' in filename) & ('taz_summaries_2010' not in filename):
        print(filename)
        yr = filename.split('.')[0][-4:]
        df = pd.read_csv(os.path.join(runs_dir, filename), usecols=['TAZ', 'TOTHH', 'TOTEMP'])
        df.rename(columns = {'TOTHH':  'tothh_'+yr,
                             'TOTEMP': 'totemp_'+yr}, inplace=True)
#         display(df.head())
        np_taz = np_taz.merge(df, on='TAZ', how='outer')

np_taz.shape

for col in list(np_taz):
    if col not in ['SUPERD', 'TAZ']:
        np_taz[col] = np_taz[col].fillna(0)

display(np_taz.head())
```

```python
p_taz = TAZ_SD.copy()
for filename in os.listdir(runs_dir):
    if ('run313_taz_summaries' in filename) & ('taz_summaries_2010' not in filename):
        print(filename)
        yr = filename.split('.')[0][-4:]
        df = pd.read_csv(os.path.join(runs_dir, filename), usecols=['TAZ', 'TOTHH', 'TOTEMP'])
        df.rename(columns = {'TOTHH':  'tothh_'+yr,
                             'TOTEMP': 'totemp_'+yr}, inplace=True)
#         display(df.head())
        p_taz = p_taz.merge(df, on='TAZ', how='outer')

p_taz.shape

for col in list(p_taz):
    if col not in ['SUPERD', 'TAZ']:
        p_taz[col] = p_taz[col].fillna(0)

display(p_taz.head())
```

```python
# wide table

for i in ['hh', 'emp']:
    for j in ['2020', '2025', '2030', '2035', '2040', '2045', '2050']:
        np_taz['growth_'+i+'_'+j+'_np'] = np_taz['tot'+i+'_'+j] - np_taz['tot'+i+'_2015']

for i in ['hh', 'emp']:
    for j in ['2020', '2025', '2030', '2035', '2040', '2045', '2050']:
        p_taz['growth_'+i+'_'+j+'_p'] = p_taz['tot'+i+'_'+j] - p_taz['tot'+i+'_2015']

# merge
taz_growth_comp = np_taz.merge(p_taz, on=['SD_NAME', 'SUPERD', 'TAZ'], how='outer')
print(taz_growth_comp.shape)

# drop unneeded columns
growth_cols = [col for col in taz_growth_comp.columns if 'growth' in col]
taz_growth_comp = taz_growth_comp[['SD_NAME', 'SUPERD', 'TAZ']+growth_cols]
```

```python
for i in ['hh', 'emp']:
    for j in ['2020', '2025', '2030', '2035', '2040', '2045', '2050']:
#         print(i+'_growth_diff_'+j)
#         print('growth_'+i+'_'+j+'_p')
#         print('growth_'+i+'_'+j+'_np')
        taz_growth_comp[i+'_growth_diff_'+j] = taz_growth_comp['growth_'+i+'_'+j+'_p'] - taz_growth_comp['growth_'+i+'_'+j+'_np']
```

```python
taz_growth_comp.to_csv(os.path.join(output_dir, 'taz_growth_comp_allYrs.csv'), index=False)
```

```python
growth_raw_cols = [col for col in taz_growth_comp.columns if ('growth_diff' not in col) and ('growth' in col) ]

growth_raw_cols
```

```python
taz_growth_stack = pd.DataFrame(columns=['SD_NAME', 'SUPERD', 'TAZ', 'growth', 'type', 'year', 'scenario'])
for i in growth_raw_cols:
    print(i)
    ty = i.split('_')[1]
    print(ty)
    yr = i.split('_')[2]
    print(yr)
    scn = i.split('_')[3]
    print(scn)
    
    df = taz_growth_comp[['SD_NAME', 'SUPERD', 'TAZ', i]]
    df.rename(columns={i: 'growth'}, inplace=True)
    df['type'] = ty
    df['year'] = yr
    df['scenario'] = scn
    display(df)
    
    taz_growth_stack = pd.concat([taz_growth_stack, df])
```

```python
taz_growth_stack.to_csv(os.path.join(output_dir, 'taz_growth_comp_allYrs_stack.csv'), index=False)
```
